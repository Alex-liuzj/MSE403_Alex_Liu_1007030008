### <span style="color: red"> Task 2:</span> Who has the most runner-up (i.e., second place) finishes in the histroy of NHL? Create a table with the teams and their runner-up finishes. The table must contain the the following columns:

-Team

-Total number of runner-up finishes

-Years [list of the years they finished as a runner-up]

You have to use Jupyter for the practice and produce a notebook with good readability.

### <span style="color: blue"> Solution:</span>

```python
import pandas as pd
import numpy as np

file = 'Stanely_cup_winners_by_year.xlsx'
df = pd.read_excel(file)  #read the excel file into dataframe


i = 0
dict2 = {'Team':[],'Runner_up_finishes':[],'Years':[]} #Create a dictionary with keys(columns)

while i < len(df):
    team_w = df['Team'][i]
    team_r = df['Runner up'][i]
    year = df['Year'][i]   #in each row, read the values of dataframe
    
    if team_r not in dict2['Team']:
        dict2['Team'].append(team_r)   #put the team in dictionary
        dict2['Years'].append([year])   #put the Runner-up year in dictionary
        dict2['Runner_up_finishes'].append(1)   #the number of Runner-up is 1
    
    else:
        indx = dict2['Team'].index(team_r) 
        dict2['Years'][indx].append(year)   #put another Runner-up year in the dictionary
        dict2['Runner_up_finishes'][indx] += 1   #add 1 to the number of Runner-up
      
    i+=1 

df2 = pd.DataFrame(dict2)   #convert the dictionary to dataframe
df2 = df2.sort_values(by='Runner_up_finishes',ascending=False)   #sort the dataframe based on number of Runner-up finishes
df2 = df2.set_index(np.arange(0,len(df2)))

print(df2['Team'][0]+' has the most runner-up finishes in the history of NHL, with '+ str(df2['Runner_up_finishes'][0])+
      ' runner-up finishes in ' + str(df2['Years'][0]))

df2
```

```python
import matminer
import pandas
from matminer.datasets import load_dataset

df = load_dataset('steel_strength')   #load the dataset 'steel_strength'

df.describe()   #statistics check

from matminer.featurizers.conversions import StrToComposition
from tqdm.autonotebook import tqdm

df = StrToComposition().featurize_dataframe(df, "formula")   #Featurize the composition

blacklist = []

df2=df.select_dtypes(include=['float64'])  #Exclude the columns where the values are not float numbers
for column in df2:
    for column2 in df2:
        if column not in blacklist:   #When two columns have high correlations, avoid dropping both of them
            if abs(df2[column].corr(df2[column2])) > 0.8 and column != column2:
                df = df.drop([column2], axis=1)   #Drop the columns with correlation coefficient greater than abs(0.8)
                blacklist.append(column2)   #After a column has been dropped, avoid iterating it again
            
df
  
    




```

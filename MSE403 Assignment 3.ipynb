```python
#Assignment 1 Task 5

import re
from mendeleev import element
def normalize(alloy):
    
    #seperate the letters and numbers
    letters = re.findall('[A-Z][^A-Z]*', alloy)  #copilot
    #put all the float number and integer number in a list
    numbers = re.findall(r'\d+\.\d+|\d+', alloy)  #copilot

    for n in numbers:
        letters = [re.sub(n, '', i) for i in letters]

    #sum up the numbers
    numbers = [float(i) for i in numbers] #copilot
    sum_num = sum(numbers)   #copilot
    
    try:
        element(letters)
    
        if sum_num == 1:
            print('the alloy\'s composition is normalized')
        else:
            normalized_number = [str(round(i/sum_num,2)) for i in numbers]  #copilot
            for i in range(len(letters)):
                letters[i] = letters[i] + normalized_number[i]   #copilot
            print('The composition of the alloy is not normalized. The normalized composition is: ' + ''.join(letters))
        
    except:
        print('There is non-elements in the formula')
```

```python
import pandas as pd
import numpy as np

file = 'Stanely_cup_winners_by_year.xlsx'
df = pd.read_excel(file)

i=0
dict2 = {'Team':[],'Runner_up_finishes':[],'Years':[]} # create a dictionary to store the data

#read the data from df
while i < len(df):
    if df['Runner up'][i] not in dict2['Team']:
        dict2['Team'].append(df['Runner up'][i])  #copilot
        dict2['Runner_up_finishes'].append(1)
        dict2['Years'].append([df['Year'][i]])    #copilot
        i += 1
    else:
        index = dict2['Team'].index(df['Runner up'][i])   #I copied my assignment 2 code to another file in pycharm, and then copilot seems to know what I want to do
        dict2['Runner_up_finishes'][index] += 1      #copilot
        dict2['Years'][index].append(df['Year'][i])   #copilot
        i += 1

# create a dataframe to store the data
df2 = pd.DataFrame(dict2)   #copilot

# sort the dataframe by the number of runner up finishes
df2 = df2.sort_values(by='Runner_up_finishes',ascending=False)   # copilot

#reset the index
df2 = df2.reset_index(drop=True)    #copilot

print(df2['Team'][0]+' has the most runner-up finishes in the history of NHL, with '+ str(df2['Runner_up_finishes'][0])+
      ' runner-up finishes in ' + str(df2['Years'][0]))

# print the dataframe
print(df2)  #copilot



```
